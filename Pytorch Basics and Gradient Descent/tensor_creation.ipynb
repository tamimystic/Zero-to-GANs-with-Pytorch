# PyTorch Tensor Creation Functions

## A. Tensor Creation (Basic Level)

These functions are used to create tensors in PyTorch.

| Function           | Purpose                                                                  |
| ------------------ | ------------------------------------------------------------------------ |
| `torch.tensor()`   | Converts normal data into a tensor                                       |
| `torch.zeros()`    | Creates a tensor filled with zeros                                       |
| `torch.ones()`     | Creates a tensor filled with ones                                        |
| `torch.empty()`    | Allocates memory without initialization (contains random/garbage values) |
| `torch.rand()`     | Creates random values between `0–1`                                      |
| `torch.randn()`    | Creates random values from a normal distribution                         |
| `torch.randint()`  | Creates random integer values                                            |
| `torch.arange()`   | Creates a range of values                                                |
| `torch.linspace()` | Creates equally spaced values                                            |
| `torch.eye()`      | Creates an identity matrix                                               |

# Key Differences

## `torch.rand()` vs `torch.randn()`

| Function        | Output Type                                                    |
| --------------- | -------------------------------------------------------------- |
| `torch.rand()`  | Random values between `0–1`                                    |
| `torch.randn()` | Random values from a normal distribution (positive + negative) |

---

## `torch.arange()` vs `torch.linspace()`

| Function           | Purpose                                         |
| ------------------ | ----------------------------------------------- |
| `torch.arange()`   | Creates values with a fixed step size           |
| `torch.linspace()` | Creates a fixed number of equally spaced values |

---

# Beginner Notes

* `torch.tensor()` → Convert your own data into a tensor
* `torch.zeros()` → Initialize tensors with zeros
* `torch.ones()` → Use for masks or default values
* `torch.empty()` → Avoid as a beginner (contains uninitialized values)
* `torch.rand()` → Generate random values between `0–1`
* `torch.randn()` → Commonly used for weight initialization
* `torch.randint()` → Generate random integer labels
* `torch.arange()` → Create ranges of numbers
* `torch.linspace()` → Create equally spaced values
* `torch.eye()` → Create identity matrices
