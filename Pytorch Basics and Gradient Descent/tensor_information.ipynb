# PyTorch Tensor Information Functions

## B. Tensor Information (Basic Level)

These functions are used to get information about a tensor.

They help answer questions like:

* What is the tensor shape?
* How many dimensions does it have?
* What is the data type?
* Is it on CPU or GPU?
* How many total elements exist?

---

# 1. `.shape`

Returns the **shape (structure)** of a tensor.

### Example

```python id="mh3m9v"
import torch

x = torch.tensor([
    [1, 2, 3],
    [4, 5, 6]
])

print(x.shape)
```

### Output

```python id="luij7r"
torch.Size([2, 3])
```

Meaning:

* 2 rows
* 3 columns

---

# 2. `.size()`

Returns the **size of the tensor**.

It works similarly to `.shape`.

### Example

```python id="jlwm8f"
x = torch.rand((2, 4))

print(x.size())
```

### Output

```python id="18t6gh"
torch.Size([2, 4])
```

---

### Get Specific Dimension Size

```python id="w31t7r"
print(x.size(0))
```

Output:

```python id="kz6bgz"
2
```

```python id="z09d6o"
print(x.size(1))
```

Output:

```python id="s74y10"
4
```

---

# 3. `.dtype`

Returns the **data type** of a tensor.

### Example

```python id="2zvucq"
x = torch.tensor([1, 2, 3])

print(x.dtype)
```

### Output

```python id="if1gcf"
torch.int64
```

---

### Float Example

```python id="m8oowu"
x = torch.tensor([1.0, 2.0])

print(x.dtype)
```

### Output

```python id="s4vhuk"
torch.float32
```

---

### Common Data Types

| Type            | Meaning                |
| --------------- | ---------------------- |
| `torch.int32`   | Integer                |
| `torch.int64`   | Large Integer          |
| `torch.float32` | Float                  |
| `torch.float64` | Double Precision Float |
| `torch.bool`    | True / False           |

---

### Type Conversion

```python id="cknpju"
x = x.float()
```

```python id="jlwmjl"
x = x.int()
```

# 4. `.device`

Returns the **device location** of a tensor.

It tells whether the tensor is stored on:

* CPU
* GPU

### Example

```python id="st7kz0"
x = torch.tensor([1, 2, 3])

print(x.device)
```

### Output

```python id="2ksjlwm"
cpu
```

---

### GPU Example

```python id="1sllgw"
device = torch.device("cuda")

x = x.to(device)

print(x.device)
```

### Output

```python id="s7flsh"
cuda:0
```

---

# 5. `.ndim`

Returns the **number of dimensions** of a tensor.

### Example 1 — Scalar

```python id="zr5t5f"

print(x.ndim)
```

Output:

```python id="k7ojxk"
0
```

---

### Example 2 — Vector

```python id="br7v1h"
x = torch.tensor([1, 2, 3])

print(x.ndim)
```

Output:

```python id="lttowg"
1
```

---

### Example 3 — Matrix

```python id="vjqkj6"
x = torch.tensor([
    [1, 2],
    [3, 4]
])

print(x.ndim)
```

Output:

```python id="mnf9kc"

```

# 6. `.numel()`

Returns the **total number of elements** in a tensor.

### Example

```python id="c4g0g2"
x = torch.tensor([
    [1, 2, 3],
    [4, 5, 6]
])

print(x.numel())
```

### Output

```python id="psuzxr"
6
```

Because:

```python id="dppnm9"
2 × 3 = 6
```

---

### Another Example

```python id="75j5kp"
x = torch.rand((4, 5, 2))

print(x.numel())
```

### Output

```python id="udl6k0"
40
```

Because:

```python id="fz6t6v"
4 × 5 × 2 = 40
```
