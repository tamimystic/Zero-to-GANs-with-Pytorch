### Tensor Operations

```python
x= torch.tensor(3.)
w=torch.tensor(4., requires_grad=True)
b=torch.tensor(5.,requires_grad=True)
x,w,b
```

```python
y = w*x+b
y
```

```python
y.backward()
```

```python
y
```

```python
print(x)
print(w)
print(b)
```

```python
print("dy/dw = x", w.grad)
print("dy/dx = w", x.grad)
print("dy/db = 1", b.grad)
```

here `x.grad = none`, beacause we did not declare `x` as `requires_grad=true`
